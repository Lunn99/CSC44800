```python
import random
import matplotlib.pyplot as plt
```

```python
def rollDice():
    roll = random.randint(1,100)
    if roll == 100:
        return False
    elif roll <= 50:
        return False
    elif 100 >roll >= 50:
        return True
```

```python
def simple_bettor(funds,wager,total_round):
    remains = funds
    wX = []
    vY = []
    current_round = 1
    while(current_round <= total_round ):
        if rollDice():
            remains += wager
            wX.append(current_round)
            vY.append(remains)
        else:
            remains -= wager
            wX.append(current_round)
            vY.append(remains)
        current_round += 1
    plt.plot(wX,vY)
    if remains <= 0:
        return "broke"
    elif remains < funds:
        return 'lose'
    elif remains > funds:
        return 'win'
    elif remains == funds:
        return 'equal'
        
```

```python
def create_model(people,funds,wager,rounds):
    num_win = 0
    num_lose = 0
    num_broke = 0
    num_equal = 0
    for i in range(people):
        result = simple_bettor(funds,wager,rounds)
        if result == 'win':
            num_win += 1
        if result == 'lose':
            num_lose += 1
        if result == 'broke':
            num_lose += 1
            num_broke += 1
        if result == 'equal':
            num_equal += 1

    plt.ylabel('Remaining Funds')
    plt.xlabel('Number of round')
    plt.show()

    print('Number of people:', people)
    print('Number of game:', rounds)
    print('Num_win:', num_win,'\tWin rate:', num_win/people*100,'%')
    print('Num_lose:', num_lose,'\tLose rate:', num_lose/people*100,'%')
    print('Num_equal:', num_equal)
    print('Num_broke:', num_broke,'\tBroken rate:', num_broke/people*100,'%')
```

```python
create_model(100,10000,100,100)
```

```python
create_model(100,10000,100,30)
```

```python
create_model(100,10000,100,1000)
```
